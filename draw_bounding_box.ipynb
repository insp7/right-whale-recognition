```python
import pandas as pd
import cv2
import os

# Function to draw bounding boxes on images
def draw_bounding_boxes(csv_file_path, image_folder, output_folder):
    # Read the CSV file
    df = pd.read_csv(csv_file_path)

    # Iterate over each row in the dataframe
    for index, row in df.iterrows():
        image_name = row['Image']
        whale_id = row['whaleID']
        x, y, width, height = int(row['x']), int(row['y']), int(row['width']), int(row['height'])

        # Read the image
        image_path = os.path.join(image_folder, image_name)
        print(f"Loading image: {image_path}")
        image = cv2.imread(image_path)

        if image is None:
            print(f"Error: Failed to load image {image_path}")
            continue

        # Draw bounding box on the image
        cv2.rectangle(image, (x, y), (x + width, y + height), (0, 255, 0), 2)

        # Save the image with bounding box
        output_path = os.path.join(output_folder, f"{image_name}")
        cv2.imwrite(output_path, image)

        print(f"Bounding box drawn for {whale_id} in image {image_name}. Saved as {output_path}")
```

```python
# Specify the path to the CSV file, the folder containing images, and output folder
csv_file_path = "annotations/whale_face_bounding_box_coordinates.csv"
image_folder = "C:\\Users\\Aniket Konkar\\Desktop\\RightWhaleRecognition\\dataset\\imgs"
output_folder = "C:\\Users\\Aniket Konkar\\Desktop\\RightWhaleRecognition\\dataset\\imgs_head_localized"

# Call the function to draw bounding boxes on images
draw_bounding_boxes(csv_file_path, image_folder, output_folder)
```
