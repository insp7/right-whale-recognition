```python
import pandas as pd
import cv2
import os

# Function to extract bounding boxes and save them as images
def extract_bounding_boxes(csv_file_path, image_folder, output_folder):
    # Read the CSV file
    df = pd.read_csv(csv_file_path)

    # Iterate over each row in the dataframe
    for index, row in df.iterrows():
        image_name = row['Image']
        whale_id = row['whaleID']
        x, y, width, height = int(row['x']), int(row['y']), int(row['width']), int(row['height'])

        # Read the image
        image_path = os.path.join(image_folder, image_name)
        image = cv2.imread(image_path)

        # Extract bounding box
        bounding_box = image[y:y+height, x:x+width]
        
        # Save the bounding box as a separate image
        output_path = os.path.join(output_folder, f"{image_name}")
        cv2.imwrite(output_path, bounding_box)

        print(f"Bounding box saved for {whale_id} in image {image_name}.")
```

```python
# Specify the path to the CSV file and the folder containing images
csv_file_path = "annotations/whale_face_bounding_box_coordinates.csv"
image_folder = "C:\\Users\\Aniket Konkar\\Desktop\\RightWhaleRecognition\\dataset\\imgs"
output_folder = "C:\\Users\\Aniket Konkar\\Desktop\\RightWhaleRecognition\\dataset\\imgs_whale_heads_only"

# Call the function to extract bounding boxes
extract_bounding_boxes(csv_file_path, image_folder, output_folder)
```
